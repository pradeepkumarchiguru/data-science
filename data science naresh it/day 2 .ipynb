# immunitable
- float , int , string , boolean  these are immunitable 
- we cannot modify the value 

```python
x=10
print("x=",x)
print("location in memory :", id(x))
```

```python
x=45
print("x=",x)
print("location of x:",id(x))
```

- here the object which we created had been deleted and new object has been created x=45 at different location 

```python
x=45
y=45
print(x,"=",id(x))
print(y,"=",id(y))
```

- keep this in u r mind object having same value will have same location of memory
